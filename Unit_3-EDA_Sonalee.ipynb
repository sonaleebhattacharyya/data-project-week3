#### Fill in blanks with a brief description of each plot type:

  
- Histogram
> A histogram is a bar graph that represents the frequency distribution for continuous data. Each bar height represents the number of observations within the specified range. 
- Density Plot / KDE
> A smooth representation of the distribution of data. It shows the estimated probability density function of the given random variable. 
- Bar Chart
> A bar chart is similar to a histogram except it represents categorical data. The bars have space between them. The height of the bars represents the number of observations in the corresponding category. 
- Line Plot
> A line plot is a graph that displays data along a number line by connecting data points with a line. It is used to show trends over time or the relationship between two variables. 
- Univariate scatter plot
> A scatter plot which displays the relationship between one independent variable and one dependent variable. It is used to identify patterns and outliers in the data. 
- Violin Plot
> A combination of a box plot and a density plot that shows the distribution of data. It is used to represent the probability density of the data at different values. 
- Box Plot
> A graph that shows the summary of the set of data values including minimum, first quartile, median, third quartile, and maximum. 
- Pie Chart
> A pie chart is a circular graph that shows the proportions of different categories of data. It is used to represent the proportion of each category relative to the whole. 
- Rug Plot
> A plot that displays individual data points along a number line. It is used to display the density of observations in a given dataset.

## Part 2 - Summary Statistics

### Step 1 - Review Summary Statistics

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

- mean
> average (sum all data points and divide by total number) 
- median
> middle number of a data set. If there are an even number of observations, average the middle two. 
- mode
> Number that occurs the most. There can be no mode or multiple modes. If every data point occurs the same number of times, there is no mode. 
- central tendency
> A descriptive summary of a dataset through a single value that reflects the center of the distribution. 
- distribution
> Shows the possible values of a variable and how often they occur. 
- variance
> The population and sample variance are found by squaring the standard deviation, or summing the squares of the differences between each data value and the mean and dividing by the number of data points (n-1 in the case of the sample variance). This is a measure of variability in the distribution. 
- standard deviation
> The standard deviation is the square root of the variance. It also measures variability in the distribution. A small standard deviation indicates the values are clustered about the mean. A larger standard deviation indicates they are more spread out.

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

<h3 style='color:red;'> First, load 'salary_numeric.csv' - where education has been re-expressed as numeric variables.</h3>

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### I hypothesize that higher education level leads to higher salary. In this case the null hypothesis would be that higher education does not lead to higher salary. 

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df['salary'], df['education'])
```

There is a positive correlation between education level and salary. 

#### Are these results statistically significant? How do you know?

Yes, typically we say that results are significant when p < .05. This means that the chance of getting an observation this extreme or greater given that the null hypothesis is true (there is no relationship between higher education and higher salary) is very small (0), so there probably is a relationship. Note it is not actually 0 of course. It is rounding. 

### What is the correlation between salary and age?

#### I hypothesize that salary increases with age. 

```python
# calculate correlation and pvalue of salary and age with pearsonr method
spearmanr(df['salary'], df['age'])
```

There is a positive correlation between salary and age. 

#### Are these results statistically significant? How do you know? 

Yes, the p value is 0. 

### What is the correlation between salary and industry experience? 

#### I hypothesize that salary will increase with industry experience. 

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
spearmanr(df['salary'], df['industry_experience'])
```

There is a positive correlation between industry experience and age. 

#### Are these results statistically significant? How do you know?

Yes, the p value is 0.
